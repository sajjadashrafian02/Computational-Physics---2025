# Excersice 6.9(generate random number between 0 and 9)

```python
import numpy as np  # Import NumPy for numerical operations
import matplotlib.pyplot as plt  # Import Matplotlib for plotting

data = []  # List to store generated numbers
N = 10000  # Total number of random numbers to generate
numbers = {}  # Dictionary to count occurrences of each number
bin=10
# Initialize the dictionary with numbers 0-9 and set count to 0
for i in range(10):
    numbers[i] = 0  

# Generate random numbers and count their occurrences
for i in range(N):
    j = np.random.randint(0, 10)  # Generate a random integer between 0 and 9
    data.append(j)  # Store in the list
    numbers[j] += 1  # Increment count in the dictionary

# Plot the histogram
plt.hist(data, bins=10, color='blue', edgecolor='black')
# Display the histogram
plt.show()

# Print the count of each number
print(numbers)
```

```python
import numpy as np  # Import NumPy for numerical operations
import matplotlib.pyplot as plt  # Import Matplotlib for plotting

# Loop through values of i from 1 to 9999 with a step of 10
for i in range(1, 10000, 10):  
    count = 0  # Initialize count of occurrences of 0

    # Generate 'i' random numbers and count occurrences of 0
    for j in range(i):
        k = np.random.randint(0, 10)  # Generate a random number between 0 and 9
        if k == 0:  # Check if the number is 0
            count += 1  # Increment count if k is 0

    # Plot scatter point for deviation from expected frequency (1/10)
    plt.scatter(i, abs(count - (i / 10)) / i, c='red')  

# Generate x values for the theoretical curve y = 1/sqrt(x)
x = np.linspace(1, 10000, 10000)  
y = 1 / np.sqrt(x)  # Compute corresponding y values

# Plot the theoretical function y = 1/sqrt(N)
plt.plot(x, y, color='b')  

# Add title to the plot
plt.title('Plot of $y = \\frac{1}{\sqrt{N}}$')

# Display the plot
plt.show()
```

# EXcersice(taking numbers just after 4)

```python
import numpy as np  # Import NumPy for numerical operations
import matplotlib.pyplot as plt  # Import Matplotlib for plotting

# Initialize variables
data = []  # List to store generated random numbers
N = 10000  # Number of random numbers to generate
numbers = {}  # Dictionary to count occurrences of numbers appearing after 4
original_data = []  # List to store numbers appearing after a 4

# Initialize dictionary with numbers 0-9 and set counts to 0
for i in range(10):
    numbers[i] = 0  

# Generate N random numbers between 0 and 9
for i in range(N):
    j = np.random.randint(0, 10)  # Generate a random integer between 0 and 9
    data.append(j)  # Append to data list

# Check for occurrences of numbers appearing after a 4
for k in range(len(data) - 1):  # Iterate through data except the last element
    if data[k] == 4:  # If the current number is 4
        original_data.append(data[k + 1])  # Store the next number
        numbers[data[k + 1]] += 1  # Increment count in dictionary

# Plot histogram of all generated numbers
plt.hist(data, bins=10, color='blue', edgecolor='black')

# Display the histogram
plt.show()

# Print count of numbers that appeared after a 4
print(numbers)
```

# generate Pseudorandom with  Modular arithmetic 


```python
def Modular_arithmetic(N):  
    m = pow(2, 31)  # Modulus value (2^31)
    c = 1013904223  # Increment value
    a = 1664525  # Multiplier value
    x = 4  # Initial seed value
    numbers = []  # List to store generated numbers

    # Generate N numbers using the linear congruential generator (LCG) formula
    for i in range(N):
        numbers.append(x)  # Store current value of x
        x = (a * x + c) % m  # Update x using the LCG formula

    return numbers  # Return the list of generated numbers

# Call the function with N=10 and print the result
print(Modular_arithmetic(10))
```

```python
import numpy as np  # Import NumPy (not used directly here but useful for future enhancements)
import matplotlib.pyplot as plt  # Import Matplotlib for plotting

# Function to generate pseudo-random numbers mod 10 using LCG (Linear Congruential Generator)
def Modular_arithmetic10(N):  
    m = pow(2, 31)  # Modulus value (2^31)
    c = 1013904223  # Increment value
    a = 1664525  # Multiplier value
    x = 4  # Initial seed value
    numbers = []  # List to store generated numbers

    # Generate N numbers using the LCG formula
    for i in range(N):
        numbers.append(x % 10)  # Store last digit (mod 10)
        x = (a * x + c) % m  # Update x using LCG recurrence relation

    return numbers  # Return the list of generated numbers

# Generate 10,000 numbers and store them in 'data'
data = Modular_arithmetic10(10000)

# Plot histogram of the generated numbers
plt.hist(data, bins=10, color='blue', edgecolor='black')

# Display the histogram
plt.show()
```

# Excersice 6.5(Central Limit Theorem (CLT))

## N=5

```python
import numpy as np  # Import NumPy for random number generation
import matplotlib.pyplot as plt  # Import Matplotlib for plotting

# Set the number of random values to average
N, numbers = 5, []  

# Generate 10,000 samples of averaged random numbers
for i in range(10000):
    y = 0  # Initialize sum

    # Generate N random numbers between 0 and 9 and compute their sum
    for i in range(N):
        x = np.random.randint(0, 10)  # Random integer from 0 to 9
        y += x  # Add to sum

    # Store the average value of the N random numbers
    numbers.append(y / N)  

# Plot histogram of averaged numbers
plt.hist(numbers, bins=10, color='blue', edgecolor='black')

# Display the histogram
plt.show()
```

# Excersice 6.5(using transformation function)

```python
import numpy as np  # Import NumPy for random number generation
import matplotlib.pyplot as plt  # Import Matplotlib for plotting

# Define the value for sigma^2
sigma2 = 25
data = []  # List to store the resulting data points

# Generate 10,000 samples based on the given conditions
for i in range(10000):
    x1 = np.random.randint(0, 10)  # Generate a random integer x1 between 0 and 9
    x2 = np.random.randint(0, 10)  # Generate a random integer x2 between 0 and 9
    
    if x1 > x2:  # If x1 is greater than x2
        A = np.tan(x1 ** 2)  # Calculate A as tan(x1^2)
        B = 2 * sigma2 * np.log(1 - x2 / x1)  # Calculate B using the formula
        y1 = np.sqrt(-B / (1 + A ** 2))  # Calculate y1
        y2 = y1 * A  # Calculate y2
        data.append(y1)  # Add y1 to the data list
        data.append(y2)  # Add y2 to the data list
    
    if x2 > x1:  # If x2 is greater than x1
        A = np.tan(x2 ** 2)  # Calculate A as tan(x2^2)
        B = 2 * sigma2 * np.log(1 - x1 / x2)  # Calculate B using the formula
        y1 = np.sqrt(-B / (1 + A ** 2))  # Calculate y1
        y2 = y1 * A  # Calculate y2
        data.append(y1)  # Add y1 to the data list
        data.append(y2)  # Add y2 to the data list

# Plot histogram of the generated data
plt.hist(data, bins=10, color='blue', edgecolor='black')

# Display the histogram
plt.show()
```
