# Euler algorithm (Instability)

```python
v=5  # Initial voltage
h=0.01  # Step size
T=1  # Total simulation time
V=[]  # List to store voltage values from Euler's method
t=[]  # List to store time values

def f(V):
    return -V  # Differential equation: dV/dt = -V (RC = 1)

for i in range(0,int(T/h),2):  # Loop through time steps
    v+=h*f(v)  
    v+=h*f(v)   # Euler update: V(n+1) = V(n-1) + h * f(V(n))
    print('in t==',h*i,'v=',v)  # Print current time and voltage
    V.append(v)  # Store voltage value
    t.append(h*i)  # Store corresponding time

plt.plot(t,V,label='Euler methos')  # Plot numerical solution with label

plt.xlabel('time')  # Label for x-axis
plt.ylabel('voltage')  # Label for y-axis

X=np.linspace(0,1,1000)  
y=np.array(5*pow(np.e,-1*X))  # Exact solution: V(t) = 5 * e^(-t)

plt.plot(X,y,label='analytic method')  # Plot analytical solution with label
plt.legend()  # Show legend with curve names
plt.show()  # Display the plot
```

# harmonic oscillator

## Euler method

```python
omega = 1.0        # Angular frequency
h = 0.01          # Time step
T = 10             # Total time

t=[]
X=[]
V=[]
E=[]
# Initial conditions
x = 1.0         # Initial position
v = 0.0         # Initial velocity

# Euler integration
for n in range(int(T/h)):
    x+=h * v
    v += -h * omega**2 * x
    X.append(x)
    V.append(v)
    t.append(n*h)
    # Energy calculation
    E.append( 0.5 * v**2 + 0.5 * (omega**2) * x**2)


# Plot results
X1=np.linspace(0,10,1000)
y1=np.cos(X1)
plt.plot(t, X, label="Position (x) euler")
plt.title("Harmonic Oscillator - Euler Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()
plt.plot(X1,y1,label="Position (x) analytic")
plt.title("Harmonic Oscillator - Euler Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()
plt.plot(X, V, label="Phase")
plt.title("Harmonic Oscillator - Phase Diagram")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.legend()
plt.show()
t1 = np.arange(0, T, h)

# Analytic solutions
x2 =  np.cos(omega * t1)
v2 = -1 * omega * np.sin(omega * t1)

# Plot phase diagram
plt.plot(x2, v2, label='Analytic Phase Curve')
plt.title("Analytic Phase Diagram - Harmonic Oscillator")
plt.xlabel("Position (x)")
plt.ylabel("Velocity (v)")
plt.axis('equal')
plt.show()
plt.figure()
plt.plot(t, E, label="Total Energy")
plt.title("Energy vs Time")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend()
plt.show()
```

## Euler-Cromer method

```python
omega = 1.0        # Angular frequency () of the harmonic oscillator
h = 0.01           # Time step for integration
T = 10             # Total simulation time

t = []             # Time values
X = []             # Position values
V = []             # Velocity values
E = []             # Total energy values

# Initial conditions
x = 1.0            # Initial position x(0)
v = 0.0            # Initial velocity v(0)

# Euler-Cromer integration loop
for n in range(int(T / h)):                   # Iterate for N = T/h steps
    v += -h * omega**2 * x                    # Update velocity first (semi-implicit method)
    x += h * v                                # Then update position using new velocity
    X.append(x)                               # Store current position
    V.append(v)                               # Store current velocity
    t.append(n * h)                           # Store current time
    E.append(0.5 * v**2 + 0.5 * omega**2 * x**2)  # Store total energy: KE + PE

# Plotting simulated position over time
X1 = np.linspace(0, 10, 1000)                 # Generate time points for analytic curve
y1 = np.cos(X1)                               # Analytic solution: x(t) = cos(t)

plt.plot(t, X, label="Position (x) Euler-Cromer")  # Plot numerical position
plt.title("Harmonic Oscillator - Euler-Cromer Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()

# Plotting analytic solution separately for reference
plt.plot(X1, y1, label="Position (x) analytic")    # Plot analytic position
plt.title("Harmonic Oscillator - Euler-Cromer Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()

# Phase diagram: velocity vs. position (numerical)
plt.plot(X, V, label="Phase")                # Phase plot: x vs. v
plt.title("Harmonic Oscillator - Phase Diagram")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')                            # Keep aspect ratio equal for circles/ellipses
plt.legend()
plt.show()

# Generate time points again (for analytic solution)
t1 = np.arange(0, T, h)

# Analytic solutions for phase diagram
x2 = np.cos(omega * t1)                      # Analytic position
v2 = -omega * np.sin(omega * t1)             # Analytic velocity

# Plot analytic phase diagram
plt.plot(x2, v2, label='Analytic Phase Curve')
plt.title("Analytic Phase Diagram - Harmonic Oscillator")
plt.xlabel("Position (x)")
plt.ylabel("Velocity (v)")
plt.axis('equal')
plt.show()

# Plot total energy over time
plt.plot(t, E, label="Total Energy")
plt.title("Energy vs Time")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend()
plt.show()
```

## leapfrog method

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 1.0                     # Angular frequency of the oscillator
h = 0.01                        # Time step size
T = 10                          # Total simulation time
X=[]                            # List to store position values
V=[]                            # List to store velocity values
E=[]                            # List to store total energy values
t=[]                            # List to store time values

# Initial conditions
x = 1.0                         # Starting position x(0)
v_half = 0.0 - 0.5 * h * omega**2 * x  # Half-step initialization of velocity (Leapfrog-style)

# Leapfrog integration
for n in range(int(T/h)):                      # Iterate over total number of steps (T/h)
    v_half += -h * omega**2 * x                # Update velocity using current position (half-step)
    x += h * v_half                            # Update position using half-step velocity
    v = v_half + 0.5 * h * (-omega**2 * x)     # Recover full-step velocity 
    X.append(x)                                # Store current position
    V.append(v)                                # Store current full-step velocity
    E.append(0.5 * v**2 + 0.5 * omega**2 * x**2)  # Store total mechanical energy: K.E. + P.E.
    t.append(n * h)                            # Store current time

# Plot position over time
plt.figure()
plt.plot(t, X, label='Position (x)')           # Plot position vs time
plt.title("Leapfrog Method - Harmonic Oscillator")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.show()

# Phase diagram
plt.figure()
plt.plot(X, V)                                 # Plot velocity vs position (phase space)
plt.title("Phase Diagram - Leapfrog Method")
plt.xlabel("Position (x)")
plt.ylabel("Velocity (v)")
plt.axis('equal')                              # Keep aspect ratio equal for circular motion
plt.show()

# Total energy
plt.figure()
plt.plot(t, E, label="Total Energy")           # Plot energy vs time to check conservation
plt.title("Total Energy Over Time - Leapfrog")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend()
plt.show()
```

## Velocity Verlet

```python
omega = 1.0        # Angular frequency
h = 0.01           # Time step
T = 10             # Total simulation time

t = [0, 1]         # Initialize time list with two placeholders (first two entries)
X = []             # Position list
V = []             # Velocity list
E = []             # Energy list

# Initial conditions
x = 1.0            # Initial position x
v = 0.0            # Initial velocity v
a = -omega**2 * x  # Initial acceleration from Hooke's law: a = -x
A = [a]            # Store initial acceleration 
X.append(x)        # Add initial position to X

# Use Taylor expansion to estimate x
x = X[0] + h * v - 0.5 * h**2 * omega**2 * X[0]
X.append(x)        # Append second position to X
V.append(v)        # Add initial velocity (v) to V

# Verlet (position-based) integration
for n in range(2, int(T/h)):
    x = 2 * X[-1] - X[-2] - h**2 * omega**2 * X[-1]  # Verlet position update formula
    X.append(x)                                     # Store next position
    t.append(n * h)                                 # Store time

# Estimate velocity using central difference
for n in range(1, int(T/h) - 1):
    V.append((X[n+1] - X[n-1]) / (2 * h))           # Velocity from central difference

# Compute total energy for each time step
for i in range(len(V)):
    E.append(0.5 * V[i]**2 + 0.5 * omega**2 * X[i]**2)  # E = KE + PE

# Plot position from Verlet integration
X1 = np.linspace(0, 10, 1000)        # Time array for plotting analytic solution
y1 = np.cos(X1)                      # Analytic solution: x(t) = cos(t)
plt.plot(t, X, label="Position (x) leapfrog")
plt.title("Harmonic Oscillator - leapfrog Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()

# Plot analytic position separately
plt.plot(X1, y1, label="Position (x) analytic")
plt.title("Harmonic Oscillator - leapfrog Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()

# Phase diagram: position vs velocity
plt.plot(X[:-1], V, label="Phase")     # Skip last X to match length of V
plt.title("Harmonic Oscillator - Phase Diagram")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')
plt.legend()
plt.show()

# Create time array for analytic solution
t1 = np.arange(0, T, h)

# Analytic solutions
x2 = np.cos(omega * t1)                 # x(t) = cos(t)
v2 = -omega * np.sin(omega * t1)        # v(t) = - sin(t)

# Plot analytic phase diagram
plt.plot(x2, v2, label='Analytic Phase Curve')
plt.title("Analytic Phase Diagram - Harmonic Oscillator")
plt.xlabel("Position (x)")
plt.ylabel("Velocity (v)")
plt.axis('equal')
plt.show()

# Plot total energy over time
plt.plot(t[:-1], E, label="Total Energy")  # Energy list is shorter than time list
plt.title("Energy vs Time")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend()
plt.show()
```

## Beeman's Method

```python
omega = 1.0        # Angular frequency ()
h = 0.01           # Time step
T = 10             # Total simulation time

t = []             # Time list
X = []             # Position list
V = []             # Velocity list
A = []             # Acceleration list
E = []             # Energy list

# Initial conditions
x = 1.0            # Initial position
v = 0.0            # Initial velocity
a = -omega**2 * x  # Initial acceleration from F = -kx = -x

X.append(x)        # Store initial position
V.append(v)        # Store initial velocity
A.append(a)        # Store initial acceleration

# Estimate x using Taylor expansion
x = x + h * v + 0.5 * h**2 * a
X.append(x)        # Store second position
a = -omega**2 * x  # Update acceleration at new position
A.append(a)        # Store new acceleration

# Estimate v using Beeman's velocity formula
v = v + 0.5 * h * (A[-1] + A[-2])
V.append(v)        # Store updated velocity

# Beeman integration loop starts from n = 1
for n in range(1, int(T/h)):
    # Predict next position using Beeman's method
    x = x + h * v + (h**2 / 6) * (4 * A[-1] - A[-2])
    a = -omega**2 * x      # Compute acceleration at predicted position
    A.append(a)            # Store new acceleration
    
    # Correct velocity using Beeman's method
    v = v + (h / 6) * (2 * A[-1] + 5 * A[-2] - A[n-3])
    
    X.append(x)            # Store updated position
    t.append(n * h)        # Store time
    V.append(v)            # Store updated velocity
    E.append(0.5 * v**2 + 0.5 * (omega**2) * x**2)  # Total energy: K.E. + P.E.

# Plot numerical position
X1 = np.linspace(0, 10, 1000)     # Time values for analytic curve
y1 = np.cos(X1)                   # Analytic solution: x(t) = cos(t)

plt.plot(t, X[:-2], label="Position (x) Euler-Cromer")  # Plot position (Beeman, label says Euler-Cromer by mistake)
plt.title("Harmonic Oscillator - Euler-Cromer Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()

# Plot analytic position
plt.plot(X1, y1, label="Position (x) analytic")
plt.title("Harmonic Oscillator - Euler-Cromer Method")
plt.xlabel("Time")
plt.ylabel("X")
plt.legend()
plt.show()

# Phase diagram: position vs. velocity
plt.plot(X, V, label="Phase")
plt.title("Harmonic Oscillator - Phase Diagram")
plt.xlabel("X")
plt.ylabel("V")
plt.axis('equal')  # Keep aspect ratio 1:1
plt.legend()
plt.show()

# Analytic solutions for comparison
t1 = np.arange(0, T, h)
x2 = np.cos(omega * t1)           # Exact position
v2 = -omega * np.sin(omega * t1)  # Exact velocity

# Plot analytic phase diagram
plt.plot(x2, v2, label='Analytic Phase Curve')
plt.title("Analytic Phase Diagram - Harmonic Oscillator")
plt.xlabel("Position (x)")
plt.ylabel("Velocity (v)")
plt.axis('equal')
plt.show()

# Plot total energy over time
plt.plot(t, E, label="Total Energy")
plt.title("Energy vs Time")
plt.xlabel("Time")
plt.ylabel("Energy")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic map function
def logistic_map(x, r):
    return 4 * r * x * (1 - x)

# Parameters
R = np.linspace(0.5, 1.0, 10000)  # r in [0.5, 1.0]
iterations = 1000
last = 200  # Only plot last will plot

x0 = 0.5  # Initial condition

plt.figure(figsize=(10, 6))

for r in R:
    x = x0
    for i in range(iterations - last):  # Burn-in
        x = logistic_map(x, r)
    x_vals = []
    for j in range(last):  # Capture final values
        x = logistic_map(x, r)
        x_vals.append(x)
    plt.plot([r] * last , x_vals, ',k' ,alpha=0.1)

plt.title("Bifurcation Diagram of Logistic Map")
plt.xlabel("Control parameter r")
plt.ylabel("x")
plt.show()
# Approximate bifurcation points (can be refined from plot)
r1 = 0.75
r2 = 0.87
r3 = 0.88
r4 = 0.90

delta_1 = (r2 - r1) / (r3 - r2)
delta_2 = (r3 - r2) / (r4 - r3)

print(f"Feigenbaum estimate 1:   {delta_1:.5f}")
print(f"Feigenbaum estimate 2:   {delta_2:.5f}")

```
