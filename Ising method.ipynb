```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
m,t=[],[]
cl=[]
k_b=1.380649*10**(-23)
energy=[]
C_V=[]
X=[]
```

# Ising model for T=100

```python
M=[]  # List to store order parameter 
T=100  # Temperature
N=50  # Size of the lattice (NxN)
A=[1,-1]  # Possible spin states
steps=10000  # Number of Monte Carlo steps

# Initialize lattice with random ±1 values
lattice=np.random.choice(A,[N,N])

E=0  # Initial energy

# Compute initial energy of the system (nearest neighbor interactions)
for i in range(lattice.shape[0]):
    for j in range(lattice.shape[0]):
        E+=lattice[i][j]*lattice[i][(j+1)%(lattice.shape[0])]  # Interaction with right neighbor
        E+=lattice[i][j]*lattice[(i+1)%(lattice.shape[0])][j]  # Interaction with bottom neighbor

# Function to compute energy change if spin at (i,j) is flipped
def energy_change(lattice,i,j):
    N=lattice.shape[0]
    left = lattice[i, (j - 1) % N]
    right = lattice[i, (j + 1) % N]
    up = lattice[(i - 1) % N, j]
    down = lattice[(i + 1) % N, j]
    delta_E = 2 * lattice[i, j] * (left + right + up + down)
    return delta_E

# Metropolis algorithm loop
for k in range(steps):
    i=np.random.randint(0,N)  # Pick random row
    j=np.random.randint(0,N)  # Pick random column

    # Metropolis criterion: flip spin with probability exp(-ΔE / T)
    if np.random.rand() < np.exp(-energy_change(lattice,i,j) / T):
        lattice[i, j] *= -1  # Flip the spin
        E+=energy_change(lattice,i,j)  # Update total energy
        energy.append(E)  # Track energy
        M.append(abs(np.average(lattice)))  # Track average order parameter

    # Visualize every 1000 steps
    if k%1000==0:
        print('step=',k)
        plt.imshow(lattice, cmap='gray')  # Show lattice configuration
        plt.title(f'Ising model at high T={T}')
        plt.axis('off')
        plt.show()

# Final visualization
print('step=',10000)
plt.imshow(lattice, cmap='gray')
plt.title(f'Ising model at high T={T}')
plt.axis('off')
plt.show()

# Record final order parameter
m.append(abs(np.sum(lattice))/(N*N)) 

t.append(T)  # Record temperature

# Compute correlation parameter
E_cl=0
for i in range(lattice.shape[0]):
    for j in range(lattice.shape[0]):
        E_cl+=lattice[i][j]*lattice[i][(j+1)%(lattice.shape[0])]

# Normalize correlation energy
E_cl=E_cl/pow(lattice.shape[0],2)

# Compute  correlation paramter
cl.append((E_cl-pow(np.average(lattice),2))/np.var(lattice))

# Compute Specific heat capacity
C_V.append(T**(-2)*np.var(energy)*k_b)

# Compute susceptibility
X.append(np.var(M)/T)  
```

# cpmputing gamma , alpha and V

```python
# Import the linear regression model from scikit-learn
from sklearn.linear_model import LinearRegression

# Create linear regression model instance
model = LinearRegression()

# Fit model to estimate gamma (susceptibility critical exponent)
# X-axis: log(|T_C - T|), Y-axis: log(susceptibility)
model.fit(np.log(abs(T_C - np.array(t))).reshape(-1, 1), np.log(X))
print("gamma=", model.coef_[0] * -1)  

# Fit model to estimate alpha (specific heat exponent)
model.fit(np.log(abs(T_C - np.array(t))).reshape(-1, 1), np.log(C_V))
print("alpha=", model.coef_[0] * -1)

# Fit model to estimate exponent V 
model.fit(np.log(abs(T_C - np.array(t))).reshape(-1, 1), np.log(np.array(cl) + 10**(-1)))
print("V=", model.coef_[0] * -1)
```
