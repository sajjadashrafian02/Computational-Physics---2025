#### step one

```python
# Define function f_6 to create a basic line plot
def f_6(x_1, x_2, y_1, y_2):
    # Calculate slope (A) between two points (x_1, y_1) and (x_2, y_2)
    A = (y_2 - y_1) / (x_2 - x_1)
    # Create NumPy arrays for x and y
    x = np.array([x_1, x_2])
    y = np.array([y_1, y_2])
    # Create new x (a_1) by dividing the original x into thirds and shifting
    a_1 = (x / 3) + x[0] - (x / 3)[0]
    # Create new y (b_1) similarly for the y values
    b_1 = y / 3 + y[0] - (y / 3)[0]
    # Plot the results for f_6
    return plt.plot(a_1, b_1)
# Define function f_7 to apply a rotation and plot a modified line
def f_7(x_1, x_2, y_1, y_2):
    # Calculate slope (A)
    A = (y_2 - y_1) / (x_2 - x_1)
    # Create NumPy arrays for x and y
    x = np.array([x_1, x_2])
    y = np.array([y_1, y_2])
    # Create new x (a_1) and y (b_1) in thirds as in f_6
    a_1 = (x / 3) + x[0] - (x / 3)[0]
    b_1 = y / 3 + y[0] - (y / 3)[0]
    # Rotation matrix for a 60 degree counterclockwise rotation
    rotation_matrix = np.array([[cos(radians(60)), sin(radians(60))], [-sin(radians(60)), cos(radians(60))]])
    # Points for rotation
    c_1 = np.array([a_1[0], b_1[0]])  # First point
    c_2 = np.array([a_1[1], b_1[1]])  # Second point
    # Apply the rotation to the second point
    c_3 = c_2 @ rotation_matrix - c_1 @ rotation_matrix
    # Adjust the first point by adding the shift based on the original x_1 value
    c_4 = np.array([x_1, c_3[0] + x_1])
    c_5 = np.array([y_1, c_3[1] + y_1])
    # New a_3 and b_3 after shifting and scaling
    a_3 = c_4 + (x_2 - x_1) / 3
    b_3 = c_5 + (x_2 - x_1) * A / 3
    # Plot the rotated and adjusted line
    return plt.plot(a_3, b_3)
# Define function f_8 to apply a rotation and plot another modified line (120 degrees)
def f_8(x_1, x_2, y_1, y_2):
    # Calculate slope (A)
    A = (y_2 - y_1) / (x_2 - x_1)
    # Create NumPy arrays for x and y
    x = np.array([x_1, x_2])
    y = np.array([y_1, y_2])
    # Create new x (a_1) and y (b_1) in thirds as in f_6
    a_1 = (x / 3) + x[0] - (x / 3)[0]
    b_1 = y / 3 + y[0] - (y / 3)[0]
    # Rotation matrix for a 120 degree counterclockwise rotation
    rotation_matrix1 = np.array([[cos(radians(120)), sin(radians(120))],[-sin(radians(120)), cos(radians(120))]])
    # Points for rotation
    d_1 = np.array([a_1[0], b_1[0]])  # First point
    d_2 = np.array([a_1[1], b_1[1]])  # Second point
    # Apply the rotation to the second point
    d_3 = d_2 @ rotation_matrix1 - d_1 @ rotation_matrix1
    # Adjust the first point by adding the shift based on the original x_1 value
    d_4 = np.array([x_1, d_3[0] + x_1])
    d_5 = np.array([y_1, d_3[1] + y_1])
    # New a_4 and b_4 after shifting and scaling
    a_4 = d_4 + (x_2 - x_1) * 2 / 3
    b_4 = d_5 + (x_2 - x_1) * A * 2 / 3
    # Plot the rotated and adjusted line
    return plt.plot(a_4, b_4)
# Define function f_9 to plot the final third line based on f_6
def f_9(x_1, x_2, y_1, y_2):
    # Calculate slope (A)
    A = (y_2 - y_1) / (x_2 - x_1)
    # Create NumPy arrays for x and y
    x = np.array([x_1, x_2])
    y = np.array([y_1, y_2])
    # Create new x (a_1) and y (b_1) in thirds as in f_6
    a_1 = (x / 3) + x[0] - (x / 3)[0]
    b_1 = y / 3 + y[0] - (y / 3)[0]
    # Create a_2 and b_2 after applying the two-thirds shift
    a_2 = a_1 + (x_2 - x_1) * 2 / 3
    b_2 = b_1 + A * (x_2 - x_1) * 2 / 3
    # Plot the adjusted line
    return plt.plot(a_2, b_2)
# Define initial points for the plot
x_1 = 0
x_2, y_1, y_2 = 12, 12, 12
# Call each function to plot the respective lines
f_6(x_1, x_2, y_1, y_2)
f_7(x_1, x_2, y_1, y_2)
f_8(x_1, x_2, y_1, y_2)
f_9(x_1, x_2, y_1, y_2)
# Set axis to be equal and display the plot
plt.axis('equal')
plt.show()
```

#### step two

```python
# Define the initial point and the shift magnitude (A)
x_1 = 0
x_2, y_1, y_2 = 12, 12, 12
A = 4  # Length for each side of the shape
# Initialize the starting point (a_1, b_1) to (x_1, y_1)
a_1 = x_1
b_1 = y_1
# Loop to create the four parts of the pattern
for i in range(4):
    # Calculate the new x and y values based on the angle and distance (A)
    # a_2 is the new x position after applying the cosine of the angle (60 * i)
    a_2 = a_1 + A * abs(cos(radians(60 * i)))
    # b_2 is the new y position based on the sine of the angle, with alternating signs
    b_2 = b_1 + A * sin(radians(60 * i)) * (-1) ** (i + 1)
    # Call the functions to plot different lines for each step (f_6, f_7, f_8, f_9)
    f_6(a_1, a_2, b_1, b_2)
    f_7(a_1, a_2, b_1, b_2)
    f_8(a_1, a_2, b_1, b_2)
    f_9(a_1, a_2, b_1, b_2)
    # Update the starting points for the next iteration
    a_1 = a_2
    b_1 = b_2
# Set the plot's axis to be equal and display the plot
plt.axis('equal')
plt.show()
```

#### step three

```python
# Initial point and parameters
x_1 = 0
x_2, y_1, y_2 = 12, 12, 12
c_1 = x_1  # Start x for the outer loop
d_1 = y_1  # Start y for the outer loop
A = 4 / 3  # Smaller step for inner loop
B = 4      # Larger step for outer loop
# Outer loop: runs 4 times to create different paths
for j in range(4):
    # Calculate the next x and y (c_2, d_2) based on the current j and B
    c_2 = c_1 + B * abs(cos(radians(60 * j)))  # X position for the outer loop
    d_2 = d_1 + B * sin(radians(60 * j)) * (-1) ** (j + 1)  # Y position for the outer loop
    # Set the starting points for the inner loop (initial a_1, b_1)
    a_1 = c_1
    b_1 = d_1
    # Inner loop: runs 4 times to create small steps based on A
    for i in range(4):
        # Special case: Adjust the step if (i==1 and j==1) or (i==2 and j==2)
        if (i == 1 and j == 1) or (i == 2 and j == 2):
            # Calculate the new x and y (a_2, b_2) with smaller step A and shifted angle
            a_2 = a_1 - A * abs(cos(radians(60 * (i + j))))  # Use negative sign for the x shift
            b_2 = b_1 - A * sin(radians(60 * (i + j))) * (-1) ** (i + 1 + j)  # Use negative for y shift
            # Call the functions to plot the lines
            f_6(a_1, a_2, b_1, b_2)
            f_7(a_1, a_2, b_1, b_2)
            f_8(a_1, a_2, b_1, b_2)
            f_9(a_1, a_2, b_1, b_2)
            # Update a_1 and b_1 for the next iteration
            a_1 = a_2
            b_1 = b_2
        else:
            # Regular case: Use positive steps with A for x and y
            a_2 = a_1 + A * abs(cos(radians(60 * (i + j))))  # X position for the step
            b_2 = b_1 + A * sin(radians(60 * (i + j))) * (-1) ** (i + 1 + j)  # Y position for the step
            # Call the functions to plot the lines
            f_6(a_1, a_2, b_1, b_2)
            f_7(a_1, a_2, b_1, b_2)
            f_8(a_1, a_2, b_1, b_2)
            f_9(a_1, a_2, b_1, b_2)
            # Update a_1 and b_1 for the next iteration
            a_1 = a_2
            b_1 = b_2
    # Update the outer loop points (c_1, d_1) for the next iteration
    c_1 = c_2
    d_1 = d_2
# Set the plot's axis to be equal and display the plot
plt.axis('equal')
plt.show()
```

```python
# Initial starting points
a_1 = 0  # x-coordinate of the first point
step = 0  # number of iterations (set to 0 here, change to a positive number for more iterations)
a_2, b_1, b_2 = 12, 12, 12  # Coordinates of the second poinT
# Initialize the points array with the starting points
points = [[a_1, b_1], [a_2, b_2]]
new_points = []  # List to hold newly calculated points
# Loop for `step` iterations to create the fractal
for i in range(step):
    for j in range(len(points) - 1):  # Iterate over each pair of points
        x_1, y_1 = points[j]  # First point of the segment
        x_2, y_2 = points[j + 1]  # Second point of the segment
        # Calculate the points dividing the segment into thirds
        x_3 = (2 * x_1 + x_2) / 3  # 1/3 of the way from x_1 to x_2
        y_3 = (2 * y_1 + y_2) / 3  # 1/3 of the way from y_1 to y_2
        x_5 = (x_1 + 2 * x_2) / 3  # 2/3 of the way from x_1 to x_2
        y_5 = (y_1 + 2 * y_2) / 3  # 2/3 of the way from y_1 to y_2
        # Calculate the direction of the new point (x_4, y_4) using a 60-degree rotation
        dx = x_5 - x_3  # Horizontal distance between x_3 and x_5
        dy = y_5 - y_3  # Vertical distance between y_3 and y_5
        # Point x_4, y_4 is rotated 60 degrees to form the "bump" in the fractal
        x_4 = (x_3 + x_5) / 2 - math.sqrt(3) * dy / 2  # 60-degree rotation of the middle segment
        y_4 = (y_3 + y_5) / 2 + math.sqrt(3) * dx / 2  # 60-degree rotation of the middle segment
        # Add the points of the current segment to the new points list
        new_points.append([x_1, y_1])
        new_points.append([x_3, y_3])
        new_points.append([x_4, y_4])
        new_points.append([x_5, y_5])
    # Update the `points` list to include the newly calculated points
    points = new_points
    new_points = []  # Reset new_points for the next iteratioN
# Plot the final shape by connecting the points
for k in range(len(points) - 1):
    plt.plot([points[k][0], points[k + 1][0]], [points[k][1], points[k + 1][1]])
# Ensure equal scaling for both axes
plt.axis('equal')
plt.show()
```

```python

# Initialize the starting points (triangle vertices)
a_1, b_1, a_2, b_2, a_3, b_3 = 0, 0, 1, 1, 2, 0
points = [[a_1, b_1], [a_2, b_2], [a_3, b_3]]
# Number of iterations (steps)
steps = 1
# Loop through the number of steps (iterations)
for i in range(steps-1):  # In your case, steps is 1, so this loop will run once
    new_points = []  # List to store the newly generated points
    # Rotate each point of the triangle by 45 degrees using the rotation matrix
    for j in points:
        rotation_matrix1 = np.array([[cos(radians(45)), sin(radians(45))], [-sin(radians(45)), cos(radians(45))]])
        new_points.append(np.array(j) @ rotation_matrix1)  # Rotate each point and append to new_points
    # Now, apply translation to the new points (find new points by shifting)
    for k in range(len(points) - 1):
        if 2 * k + 1 < len(points):
            # Calculate the difference (dy, dx) between consecutive points
            dy = new_points[k][1] - new_points[k + 1][1]
            dx = new_points[k][0] - new_points[k + 1][0]
            # Shift the current point by (dx, dy) and append to the list
            new_points.append([new_points[k][0] + dx, new_points[k][1] + dy])
        else:
            dy = new_points[k + 1][1] - new_points[k][1]
            dx = new_points[k + 1][0] - new_points[k][0]
            # Shift the last point by (dx, dy) and append to the list
            new_points.append([new_points[-1][0] + dx, new_points[-1][1] + dy])
    points = new_points  # Update points for the next iteration
# Plot the final shape by connecting the points
for l in range(len(points) - 1):
    plt.plot([points[l][0], points[l + 1][0]], [points[l][1], points[l + 1][1]])  # Draw each line
# Ensure the aspect ratio is equal to keep the shape proportional
plt.axis('equal')
# Show the plot
plt.show()
```

## Exercise2.4(pascall triangle)

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
# Function to generate Pascal's Triangle up to 'n' rows
def pascal_triangle(n):
    # Create an empty 2D array for the triangle with all elements initialized to 0
    triangle = np.zeros((n, n), dtype=int)
    # Loop through each row
    for i in range(n):
        # Set the first element of each row to 1
        triangle[i][0] = 1  
        # Calculate the values for each row based on the previous row
        for j in range(1, i + 1):
            triangle[i][j] = triangle[i-1][j-1] + triangle[i-1][j]  
    return triangle
# Function to display the Pascal's Triangle with color-coding for odd/even numbers
def display_pascal_triangle(n):
    # Generate the Pascal's Triangle
    triangle = pascal_triangle(n)  
    # Create a figure and axis for the plot
    fig, ax = plt.subplots(figsize=(6, 6))
    # Loop through the rows and columns of the triangle
    for i in range(n):
        for j in range(i + 1):  
            # Check if the number is odd or even
            if triangle[i][j] % 2 != 0:
                color = 'blue'  # Odd numbers are blue
            else:
                color = 'red'   # Even numbers are red
            # Place the number at the appropriate position on the plot
            ax.text(j - i / 2, -i, str(triangle[i][j]), ha='center', va='center', fontsize=40, color=color)
    # Turn off axis lines and labels
    plt.axis('off')
    # Display the plot
    plt.show()
# Set the number of rows for the Pascal's Triangle
n = 15
# Display the Pascal's Triangle
display_pascal_triangle(n)
```
